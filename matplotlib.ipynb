# Import libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import datetime
import random
import seaborn as sns
import pandas as pd
```

```python
df = pd.read_csv('Diwali Sales Data.csv',encoding = 'unicode_escape')
```

```python
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
df.drop(['Status','unnamed1'],axis = 1,inplace=True)
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
#drop null value
df.dropna(inplace=True)
```

```python
#change data type
df['Amount'] = df['Amount'].astype('int')
```

```python
df['Amount'].dtypes
```

```python
df.columns
```

```python
# rename column
df.rename(columns={'Marital_Status':'Shaddi'})
```

```python
#describe method returns description of the data in the dataframe (i.e. count,std,etc)
df.describe()
```

```python
df[['Age','Orders','Amount']].describe()
```

# Exploratory Data Analysis

### Gender

```python
df.columns
```

```python
ax= sns.countplot(data = df,x='Gender')
ax.bar_label(ax.containers[0])
plt.show()
```

```python
sales_gen = df.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)

sns.barplot(x='Gender',y='Amount',data=sales_gen)
```

```python
df.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
```

# Age

```python
df.columns
```

```python
ax= sns.countplot(data = df,x='Age Group',hue='Gender')
ax.bar_label(ax.containers[0])
plt.show()
```

```python
# total amount vs age group

sales_age = df.groupby(['Age Group'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)

sns.barplot(x='Age Group',y='Amount',data=sales_age)
```

# Marital_Status

```python
ax= sns.countplot(data = df,x='Marital_Status',hue='Gender')
ax.bar_label(ax.containers[0])
plt.show()
```

# States

```python
sales_age = df.groupby(['State'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
plt.figure(figsize=(15,5))
plt.xticks(rotation = 90)
sns.barplot(x='State',y='Amount',data=sales_age)
```

```python
sales_age = df.groupby(['State'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False)
plt.figure(figsize=(15,5))
plt.xticks(rotation = 90)
sns.barplot(x='State',y='Orders',data=sales_age)
```

# Occupation

```python
df.columns
```

```python
ax= sns.countplot(data = df,x='Occupation',hue='Gender')
sns.set(rc={'figure.figsize':(20,10)})
ax.bar_label(ax.containers[0])
plt.show()
```

# occupation 

```python
sales_age = df.groupby(['Occupation'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
plt.figure(figsize=(15,5))
sns.barplot(x='Occupation',y='Amount',data=sales_age)
```

# Product_Category

```python
ax= sns.countplot(data = df,x='Product_Category')
sns.set(rc={'figure.figsize':(20,5)})
ax.bar_label(ax.containers[0])
plt.xticks(rotation = 90)
plt.show()
```

```python
sales_age = df.groupby(['Product_Category'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
plt.figure(figsize=(15,5))
plt.xticks(rotation = 60)
sns.barplot(x='Product_Category',y='Amount',data=sales_age)
```

```python
sales_age = df.groupby(['Product_ID'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(10)
plt.figure(figsize=(15,5))
sns.barplot(x='Product_ID',y='Orders',data=sales_age)
```
